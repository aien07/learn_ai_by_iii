```python
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
import csv
import bs4

HEADERS = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.116 Safari/537.36'}

table = [
    ['', '', '', '', ''],
]
# news_id = 5089692
news_id = 5100927

while (len(table) < 10001):
    url = f'https://www.cw.com.tw/article/{news_id}'
    res = rq.get(url,headers=HEADERS) #  requests  get 
    print(f'res.status_code:{res.status_code},{news_id}')
    time.sleep(3)
    if( res.status_code == 200 ):
        res.encoding='utf-8'
        soup = bs(res.text, "lxml")
        if len(soup.select('.my10'))>0:
            # 
            for s in soup.select('a'):
                s.extract()
            # 
            title = soup.select_one('.my10').text.replace('\u3000','')
            # 
            procontent = soup.select('.article__content>*:not(.ad):not(.imgzoom):not(noscript):not(script):not(address), .preface p')
            content = ''
            for c in procontent:
                # 
                if c.select('.imgzoom'):
                    continue
                else:
                    content_temp = c.text.replace('\n','').replace('\r','').replace('\xa0','').replace('\u3000','').replace('','').replace('',' ').replace('',' ').replace('',' ').replace('',' ').replace('',' ')
                    content = content + content_temp
            # 
            profigs = soup.select('.article__content>p .imgzoom img, .article__head .imgzoom .lazyload')
            figs_src = []
            for i, c in enumerate(profigs):
                if i < 1:
                    # src
                    figs_src.append(c['data-src'])
                else:
                    figs_src.append(c['src'])
            figs = '###'.join(figs_src)
            table.append([title, url, content, figs, True])
    news_id = news_id - 1
# csv
with open('cw.csv', 'w', newline='', encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)

    # 
    writer.writerows(table)
```
