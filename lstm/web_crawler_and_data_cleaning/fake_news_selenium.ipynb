```python
# V2
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
import csv
import re

path = 'C:\AIclass\project_homework\chromedriver'
options = webdriver.ChromeOptions()
# options.add_argument('--disable-notifications')

browser = webdriver.Chrome(executable_path=path,options=options)
browser.set_page_load_timeout(1000)

first_url = ["https://www.mygopen.com/search/label/%E8%AC%A0%E8%A8%80", "https://www.mygopen.com/search/label/%E8%A9%90%E9%A8%99"]

# 想抓的東西
table = [
    ['新聞標題', '網址', '內文', '照片', '真假'],
]
# 爬蟲
for url in first_url:
    browser.get(url)
    time.sleep(3)
    soup = bs(browser.page_source, 'lxml')
    # 抓總頁數
    page = soup.select('.archive-page-pagination-bottom a')
    page_url = []
    for p in page:
        page_url.append(p['href'])
    # 進每頁
    for p in page_url:
        page_go_url = 'https://www.mygopen.com/search/label/%E8%AC%A0%E8%A8%80' + p
        browser.get(page_go_url)
        time.sleep(3)
        soup = bs(browser.page_source, 'lxml')
        # 抓每篇的網址
        items = soup.select('#widget-content-HTML4 .item .item-thumbnail')
        items_url = []
        for i in items:
            items_url.append(i['href'])
        # 進要抓東西的網頁
        for url in items_url:
            browser.get(url)
            print(f'分類網址: {page_go_url}, 本篇網址: {url}')
            time.sleep(3)
            soup = bs(browser.page_source, 'lxml')
            # 抓標題
            title = re.sub(r"^【\w+】", '', soup.select_one('.post-header .post-title').text)
            # 內文
            procontent = soup.select('.tr_bq .blockquote-content')
            content = ''
            for c in procontent:
                content = content + c.text.replace('\n','')
            # 抓圖片網址
            profigs = soup.select('.post-body-inner>a>img')
            figs_src = []
            for i, c in enumerate(profigs):
                if i > 0:
                    figs_src.append(c['src'])
            figs = '###'.join(figs_src)
            # 加進資料庫
            table.append([title, url, content, figs, False])
# 寫進csv
with open('mygopen.csv', 'w', newline='', encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)

    # 寫入二維表格
    writer.writerows(table)
browser.quit()
```
