```python
df = df.assign(新聞標題_temp=lambda x: x.新聞標題, 內文_temp = lambda x: x.內文)
df
```

```python
import re
def del_title_noise(string):
    temp = re.sub(r"\w+[《]?\w*[：！》]?\w+\W?$", '', string)
    return re.sub(r"^ *【\w+】", '', temp)
df.loc[:,'新聞標題'] = df.loc[:,'新聞標題_temp'].apply(del_title_noise)
```

```python
def del_content_noise(string):
    return string.replace('\ufeff', '').replace('\u2060', '')
df.loc[:,'內文'] = df.loc[:,'內文_temp'].apply(del_content_noise)
```

```python
df = df.assign(內文_type = lambda x: x.內文)
def del_content_type(string):
    return type(string)
df.loc[:,'內文_type'] = df.loc[:,'內文_temp'].apply(del_content_type)
df
```

```python
# 簡體到繁體（臺灣正體標準）並轉換爲臺灣常用詞彙
import opencc
converter = opencc.OpenCC('s2twp.json')
def del_content_noise(string):
    string = string.replace('\ufeff', '').replace('\u2060', '')
    return converter.convert(string)
df.loc[:,'內文'] = df.loc[:,'內文_temp'].apply(del_content_noise)
```
