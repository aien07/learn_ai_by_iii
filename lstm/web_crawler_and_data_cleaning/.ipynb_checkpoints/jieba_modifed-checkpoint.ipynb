```python
# df.append(df2, ignore_index=True)
import pandas as pd

# load csv file
df_train = pd.read_csv('train_v2.csv')
df_test = pd.read_csv('test.csv')
df_total = df_train.append(df_test, ignore_index=True)
# df_total = df_total.drop(['Unnamed: 0'], axis=1)
df_total
```

```python
import jieba
jieba.set_dictionary('dict_v2.txt')  # 如果是使用繁體文字，請記得去下載繁體字典來使用
with open('stopwords_v2.txt', 'r', encoding='utf8') as f:  # 中文的停用字，我也忘記從哪裡拿到的，效果還可以，繁體字的資源真的比較少，大家將就一下吧
    stops = f.read().split('\n')
with open('stopwords_only_symbol_v2.txt', 'r', encoding='utf8') as fs:  # 中文的停用字，我也忘記從哪裡拿到的，效果還可以，繁體字的資源真的比較少，大家將就一下吧
    stops_symbol = fs.read().split('\n')

df_total_title = df_total.loc[:, ['title', 'label']].copy()
df_total_title
df_total_title['全模式_jieba'] = df_total_title['title'].apply(lambda y:' '.join([t for t in jieba.cut(str(y), cut_all=True) if t not in stops]))
df_total_title['全模式_jieba_no_stopwords'] = df_total_title['title'].apply(lambda y:' '.join([t for t in jieba.cut(str(y), cut_all=True) if t not in stops_symbol]))
df_total_title['搜索引擎模式_jieba'] = df_total_title['title'].apply(lambda y:' '.join([t for t in jieba.cut_for_search(str(y)) if t not in stops]))
df_total_title['搜索引擎模式_no_stopwords'] = df_total_title['title'].apply(lambda y:' '.join([t for t in jieba.cut_for_search(str(y)) if t not in stops_symbol]))
df_total_title['精確模式_jieba'] = df_total_title['title'].apply(lambda y:' '.join([t for t in jieba.cut(str(y), cut_all=False) if t not in stops]))
df_total_title['精確模式_no_stopwords'] = df_total_title['title'].apply(lambda y:' '.join([t for t in jieba.cut(str(y), cut_all=False) if t not in stops_symbol]))
df_total_title
```

```python
df_total_title.to_csv('total_title_split_7_3.csv', index=False)
```
