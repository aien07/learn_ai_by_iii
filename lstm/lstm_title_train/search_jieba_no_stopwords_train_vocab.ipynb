```python
import pandas as pd
import numpy as np
import tensorflow as tf
# load csv file
df_total = pd.read_csv('total_title_split_7_3.csv')
df_total
```

```python
# split data
train_data_temp,test_data_temp = df_total[:int(len(df_total)*0.7)], df_total[int(len(df_total)*0.7):]
print(len(train_data_temp),len(test_data_temp))
```

```python
check_model_data_temp = test_data_temp[int(len(test_data_temp)*1/3):int(len(test_data_temp)*2/3)]
check_model_data_temp
```

```python
test_data_pro = test_data_temp.drop(test_data_temp[int(len(test_data_temp)*1/3):int(len(test_data_temp)*2/3)].index)
test_data_pro
```

```python
train_data = train_data_temp['搜索引擎模式_no_stopwords'].values
train_target = train_data_temp['label'].values
test_data = test_data_pro['搜索引擎模式_no_stopwords'].values
test_target = test_data_pro['label'].values
check_model_data = check_model_data_temp['搜索引擎模式_no_stopwords'].values
check_model_target = check_model_data_temp['label'].values
# train_data
# train_target
# test_data
# test_target
```

```python
tf.reset_default_graph()

epochs = 15
batch_size = 64
max_sequence_length = 30
# max_sequence_length = 150
# rnn_size = [128, 64]
rnn_size = 10
embedding_size = 50
min_word_frequency = 0
# learning_rate = 0.0001
learning_rate = 0.0005

vocab_processor = tf.contrib.learn.preprocessing.VocabularyProcessor(max_sequence_length,
                                                                     min_frequency=min_word_frequency)
```

```python
train_data_pd = np.array(list(vocab_processor.fit_transform(train_data)))
test_data_pd = np.array(list(vocab_processor.transform(test_data)))
# check_model_data_pd = np.array(list(vocab_processor.transform(check_model_data)))
vocab_processor.save('search_jieba_no_stopwords_train_vocab.pickle')
```

```python
len(vocab_processor.vocabulary_)
```

```python
#隨機打亂
shuffled_ix = np.random.permutation(np.arange(len(train_target)))
x_train = train_data_pd[shuffled_ix]
y_train = train_target[shuffled_ix]
y_train  = y_train.reshape((len(y_train),))


shuffled_ix = np.random.permutation(np.arange(len(test_target)))
x_test = test_data_pd[shuffled_ix]
y_test = test_target[shuffled_ix]
y_test  = y_test.reshape((len(y_test),))

# shuffled_ix = np.random.permutation(np.arange(len(check_model_target)))
# x_check_model = check_model_data_pd[shuffled_ix]
# y_check_model = check_model_target[shuffled_ix]
# y_check_model  = y_check_model.reshape((len(y_check_model),))
vocab_size = len(vocab_processor.vocabulary_) # 有多少詞彙

tf.reset_default_graph()

# Create placeholders
x_data = tf.placeholder(tf.int32, [None, max_sequence_length])
y_output = tf.placeholder(tf.int32, [None])
dropout_keep_prob = tf.placeholder(tf.float32) #

# Create embedding
# 构造一个vocabulary_size x embedding_size的矩阵，作为embedding_mat容器，
# 有vocabulary_size个容量为embedding_size的向量，每个向量代表一个vocabulary，
# 每个向量的中的分量的值都在-1到1之间随机分布
embedding_mat = tf.Variable(tf.random_uniform([vocab_size, embedding_size], -1.0, 1.0))
# 调用tf.nn.embedding_lookup，索引与train_dataset(x_data)对应的向量，相当于用train_dataset(x_data)作为一个id，去检索矩阵中与这个id对应的embedding
embedding_output = tf.nn.embedding_lookup(embedding_mat, x_data)

# Build two layer RNN
# RNN cell composed sequentially of multiple simple cells.
# num_units = [128, 64]
# cells = [BasicLSTMCell(num_units=n) for n in num_units]
# stacked_rnn_cell = MultiRNNCell(cells)
# cells = tf.nn.rnn_cell.MultiRNNCell([tf.contrib.rnn.BasicLSTMCell(i) for i in rnn_size])
# cells = tf.nn.rnn_cell.MultiRNNCell([tf.contrib.rnn.BasicLSTMCell(rnn_size) for _ in range(2)])
cells = tf.contrib.rnn.BasicLSTMCell(rnn_size)


output, state = tf.nn.dynamic_rnn(cells, embedding_output, dtype=tf.float32)
print(f'output before handle: {output}')
output = tf.nn.dropout(output, dropout_keep_prob) #
# Get output of RNN sequence
output = tf.transpose(output, [1,0,2])
print(f'output transpose: {output}')
output = tf.unstack(output)
print(f'output after handle: {output}')
last = output[-1]

weight = tf.Variable(tf.truncated_normal([rnn_size, 2], stddev=0.1))
bias = tf.Variable(tf.constant(0.1, shape=[2]))
print(f'bias: {bias}')
logits_out = tf.matmul(last, weight) + bias
print(f'logits_out: {logits_out}')
# Loss function
losses = tf.nn.sparse_softmax_cross_entropy_with_logits(logits=logits_out, labels=y_output) # logits=float32, labels=int32
print(f'losses: {losses}')
beta = 0.01 #
# loss = tf.reduce_mean(losses+beta*tf.nn.l2_loss(weight))
loss = tf.reduce_mean(losses)
train_step = tf.train.RMSPropOptimizer(learning_rate).minimize(loss)
# train_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss)
# train_step = tf.train.AdamOptimizer(learning_rate).minimize(loss)
accuracy = tf.reduce_mean(tf.cast(tf.equal(tf.argmax(logits_out, 1), tf.cast(y_output, tf.int64)), tf.float32))
vec2prob = tf.nn.softmax(logits_out)
probability = tf.reduce_max(vec2prob, 1, name = 'probability')
# probability1 = (logits_out/tf.reduce_sum(logits_out, 1))[0]
# probability2 = tf.nn.softmax(logits_out)
ans = tf.argmax(vec2prob, 1, name = 'ans')
```

```python
saver = tf.train.Saver() # 存模型

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())

    train_loss = []
    test_loss = []
    train_accuracy = []
    test_accuracy = []
    # Start training
    for epoch in range(epochs):

        # Shuffle training data
        shuffled_ix = np.random.permutation(np.arange(len(x_train)))
        x_train = x_train[shuffled_ix]
        y_train = y_train[shuffled_ix]
        num_batches = int(len(x_train)/batch_size) + 1
        # TO DO CALCULATE GENERATIONS ExACTLY
        for i in range(num_batches):
            # Select train data
            min_ix = i * batch_size
            max_ix = np.min([len(x_train), ((i+1) * batch_size)]) # len(x_train)與((i+1) * batch_size)取最小值
            x_train_batch = x_train[min_ix:max_ix]
            y_train_batch = y_train[min_ix:max_ix]

            # Run train step
            train_dict = {x_data: x_train_batch, y_output: y_train_batch, dropout_keep_prob: 0.5} #
            sess.run(train_step, feed_dict=train_dict)
            if i % 50 == 0:
                # Run loss and accuracy for training
                temp_train_loss, temp_train_acc = sess.run([loss, accuracy], feed_dict=train_dict)
                train_loss.append(temp_train_loss)
                train_accuracy.append(temp_train_acc)
                print('Train Loss: {:.2}, Train Acc: {:.2}'.format(temp_train_loss, temp_train_acc))
#         print('Epoch: {}, Train Loss: {:.2}, Train Acc: {:.2}'.format(epoch+1, temp_train_loss, temp_train_acc))
        # Run Eval Step
        test_dict = {x_data: x_test, y_output: y_test, dropout_keep_prob: 1} #
        temp_test_loss, temp_test_acc = sess.run([loss, accuracy], feed_dict=test_dict)
        test_loss.append(temp_test_loss)
        test_accuracy.append(temp_test_acc)
        print('Epoch: {}, Test Loss: {:.2}, Test Acc: {:.2}'.format(epoch+1, temp_test_loss, temp_test_acc))
        saver.save(sess, "Saved_model/search_jieba_no_stopwords_train_vocab.ckpt")
#     test_p_and_ans = {x_data: np.array([x_test[0]]), dropout_keep_prob: 1}
#     print(sess.run(logits_out, feed_dict=test_p_and_ans))
#     print(f'logits_out: {sess.run(logits_out, feed_dict=test_p_and_ans)}')
#     print(f'probability: {probability}, {sess.run(probability, feed_dict=test_p_and_ans)}')
# #     print(f'probability1: {probability1}, {sess.run(probability1, feed_dict=test_p_and_ans)}')
# #     print(f'probability2: {probability2}, {sess.run(probability2, feed_dict=test_p_and_ans)}')
#     print(f'ans: {ans}, {sess.run(ans, feed_dict=test_p_and_ans)}')    
```

```python
epochs_train_loss = []
for i in range(int(len(train_loss)/5)):
    avg = sum(train_loss[i*5:(i+1)*5])/len(train_loss[i*5:(i+1)*5])
    epochs_train_loss.append(avg)
epochs_train_loss
```

```python
epochs_train_accuracy = []
for i in range(int(len(train_accuracy)/5)):
    avg = sum(train_accuracy[i*5:(i+1)*5])/len(train_accuracy[i*5:(i+1)*5])
    epochs_train_accuracy.append(avg)
epochs_train_accuracy
```

```python
import matplotlib.pyplot as plt
epochs_range = range(1, int(len(train_loss)/5)+1)
plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1) # (y軸要幾張圖, x軸要幾張圖, 第幾個圖)
plt.plot(epochs_range, epochs_train_accuracy, label='Training Accuracy')
plt.plot(epochs_range, test_accuracy, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, epochs_train_loss, label='Training Loss')
plt.plot(epochs_range, test_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
