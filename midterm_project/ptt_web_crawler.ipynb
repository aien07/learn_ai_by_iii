```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd

path = 'C:\AIclass\project_homework\chromedriver'
options = webdriver.ChromeOptions()
options.add_argument('--disable-notifications')

browser = webdriver.Chrome(executable_path=path,options=options)

url = "https://pttweb.tw/news"
browser.get(url)

soup = bs(browser.page_source, 'lxml')
len(soup.select('.thread-item'))

item = soup.select('.thread-item')

title, popularity, posttime = [] ,[], []
for i in item:
    
    if i.select_one('.thread-title'):
        title.append(i.select_one('.thread-title a').text)
    else:
        title.append('找不到title')
    
    if i.select_one('.thread-popularity'):
        popularity.append(i.select_one('.thread-popularity').text)
    else:
        popularity.append('找不到留言數')
    
    if i.select_one('.thread-posttime'):
        posttime.append(i.select_one('.thread-posttime').text)
    else:
        posttime.append('找不到發布時間')
                


df = pd.DataFrame({'title':title, 'popularity':popularity, 'posttime':posttime})
df = df.drop_duplicates()
df
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
import pymongo

path = 'C:\AIclass\python_url\chromedriver'
options = webdriver.ChromeOptions()
options.add_argument('--disable-notifications')

browser = webdriver.Chrome(executable_path=path,options=options)

url = "https://pttweb.tw/news"
browser.get(url)

soup = bs(browser.page_source, 'lxml')
len(soup.select('.thread-item'))

item = soup.select('.thread-item')

# title, popularity, posttime = [] ,[], []
n = 1
insert_data = []
for i in item:
    
    if i.select_one('.thread-title'):
        title = i.select_one('.thread-title a').text
    else:
        title = '找不到title'
    
    if i.select_one('.thread-popularity'):
        popularity = i.select_one('.thread-popularity').text
    else:
        popularity = '找不到留言數'
    
    if i.select_one('.thread-posttime'):
        posttime = i.select_one('.thread-posttime').text
    else:
        posttime = '0000-00-00 00:00:00'
        
    insert_data.append({"_id": n,"title":title, "popularity":popularity, "posttime":posttime})
    
    n += 1
    
myclient = pymongo.MongoClient("mongodb://localhost:27017/")

mydb = myclient["testdatabase"]

mycol = mydb["ptt_new"]

x = mycol.insert_many(insert_data)

print(x.inserted_ids)

# sqlStuff = "INSERT INTO pttweb_url (title, popularity, posttime) VALUES (%s,%s,%s)"

# cursor.executemany(sqlStuff, insert_data)
# pydb.commit()
# print(cursor.rowcount, "记录插入成功。")
```
